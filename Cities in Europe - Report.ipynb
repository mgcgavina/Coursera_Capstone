# INTRODUCTION
My friend is looking for a new city to live in Europe and asked for my help. He is from Porto, so he's looking for a similar city in terms of venues nearby. He wants a nice city that has a low living cost but with a lot of restaurants, stores and things to do.
So this project is my way of helping him. I will look for the cities similar to Porto, accordingly to the Living Cost level.


# DATA SECTION
To help me solving this problem, I will use the following data:
* Cost of Living Index of the main cities in Europe (https://www.numbeo.com/cost-of-living/region_rankings_current.jsp?region=150)
* Coordinates of World Cities Database (https://simplemaps.com/data/world-cities)
* Foursquare location data to help me with the venues

# METHODOLGY

After I've analyze my friend's problem, I've decided I would get the data from the websites selected, would analyse it and would present it in a map, as it was easier to visualize it.
So, the project was organized as follows:

1 - DATAFRAME WITH LIVING COST:
I decide to start with the Living Cost information. So my first step was to get all the information from the website Numbeo and create a Dataframe. It had more data that I've actually need it, so I cleaned it by removing the unnecessary spaces, spliting the City/Country column in two and droping the columns that we wouldn't need. I end up with a table with the basic information - City, Country and Cost of Living Index.

2 - DATAFRAME WITH COORDINATES:
The first part was ready, so I start with the second font of information that I needed for the map - coordinates.
Through the website World Cities Database I got an excel with the coordinates plus extra data. I had some troubles in importing the file from my computer, so I got the data through the Watson's Studio option of loading files directly into a dataframe.
All the information was in the dataframe so it was time to clean it - remove the unnecessary spaces (before and after) amd drop the columens we didn't need. At the end we had a table with the City and coordinates (Latitude and Longitude).

3 - BRING ALL THE DATA TOGETHER:
Both parts were ready, so it's time to bring evertything together.
I've merged both df in one, getting one dataframe with all the data we would need - City, Country, Cost of Living Index and Coordinates. i've tested it to make sure we would get the correct information and I plotted in a map.

4 - FIND THE VENUES:
We had the citys in a map, it was missing the venues. For this we use the Foursquare API to get the Top 20 Venues for each city. I've started by testing it in one city only and after I was sure it was working, I've applied to all the others. We got up with a dataframe with all the main details from the venues per city. It was a lot of data, so I've decided to sort it by the most frequent categories and get the top 10 per city.
Now it was time to separate them to get whose were similar to Porto.

5- CLUSTERING:
I set the number of clusters as 5 and run the k-means Clustering function. It was time to get all necessary information into one dataframe only, so we get it - City, Country, Cost of Living, Coordinates, Cluster Label and venues' category by order. I plotted in a map to visualize it. After that I've checked which Cluster was Porto, to get all the similar cities.

6 - FINAL STEP:
I realize it was easier to visualize it if the datapoints in the map were with different colors, accordingly to the Cost of Living Index.
So, in this step I've decided to split the Cost of Living Index data into 5 levels (bins) - Very Low, Low, Moderate, High and Very High. This information was added to the dataframe on a new column.
So, all the data was in a single dataframe - City, Country, Cost of Living, Level of Cost and top 10 venues' categories, it was time to show all the conclusions. I ploted a map center in Europe with all cities from the same cluster as Porto, with different colors as their Cost of Living level. It has the option to click on the city point and know the name and the exactly Cost of Living Index.



# RESULTS

After all the process, we came to 174 similar cities, where 63 had Very Low Cost of Living. The city with the Lowest index is Sumy in Ukraine with 27.32. The most expensive city with similar venues, is Zurich in Switzerland with index 127.69.
